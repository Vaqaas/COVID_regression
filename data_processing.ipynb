```python
import pandas as pd
import numpy as np
import math
```

```python
starter_df = pd.read_csv('data/nyimes_covid_data_by_county.csv')
len(starter_df[starter_df['Hospitalized Daily Avg'] == '—']) == len(starter_df)
```

```python
#starter_df = starter_df.drop(['Hospitalized Daily Avg'], axis=1)
starter_df.head(50)
```

```python
starter_df['Deaths Daily Avg.'] = starter_df['Deaths Daily Avg.'].replace(to_replace='—',value=np.nan)
```

```python
def p2f(x):
    if x == 0 or x == None:
        pass
    else:
        return float(x.strip('%'))
    
starter_df['Fully Vaccinated'] = starter_df['Fully Vaccinated'].replace(to_replace='—',value=0)
starter_df['Fully Vaccinated'] = starter_df['Fully Vaccinated'].map(p2f)
```

```python
starter_df['Hosp. Daily per 100k'] = starter_df['Hosp. Daily per 100k'].replace(to_replace='—',value=np.nan)
starter_df['Hosp. Daily per 100k'].unique()
```

```python
def clean_string(s):
    if type(s) != str:
        pass
    else:
        return s.replace('–', '-')

starter_df['Hosp. Daily per 100k'] = starter_df['Hosp. Daily per 100k'].map(clean_string)
```

```python
starter_df['Hosp. Daily per 100k'] = starter_df['Hosp. Daily per 100k'].replace(to_replace='Flat',value=0)
starter_df['Hosp. Daily per 100k'] = starter_df['Hosp. Daily per 100k'].map(p2f)
```

```python
starter_df['Hosp. Daily per 100k'].fillna(0, inplace=True)
starter_df['Fully Vaccinated'].fillna(0, inplace=True)
starter_df['Daily Case Avg'].replace(to_replace='<1',value=1, inplace=True)
```

```python
starter_df.replace(to_replace='—', value=pd.NA, inplace=True)
```

```python
starter_df['Deaths Daily per 100k'].fillna(-1, inplace=True)
```

```python
starter_df['Deaths Daily per 100k'] = starter_df['Deaths Daily per 100k'].map(str)
starter_df['Deaths Daily per 100k'] = starter_df['Deaths Daily per 100k'].map(p2f)
starter_df.head(50)
```

```python
ddeaths_100k_mean = starter_df['Deaths Daily per 100k'].mean()
starter_df['Deaths Daily per 100k'].replace(to_replace=-1,value=ddeaths_100k_mean, inplace=True)
starter_df.head(50)
```

```python
starter_df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
starter_df['Total Cases'] = starter_df['Total Cases'].apply(lambda x: np.float64(x.replace(',','')))
```

```python
starter_df['Total Cases per 100k'] = starter_df['Total Cases per 100k'].apply(lambda x: np.float64(x.replace(',','')))
starter_df
```

```python
starter_df['Total Cases'] = starter_df['Total Cases'].apply(lambda x: np.float64(x.replace(',','')))
```

```python
col_to_strip = ['Total Deaths','Total Deaths per 100k']
for col in col_to_strip:
        starter_df[col] = starter_df[col].apply(lambda x: x if type(x) != str else np.float64(x.replace(',','')))
```

```python
starter_df['Total Deaths'].fillna(np.nan, inplace=True)
starter_df['Total Deaths per 100k'].fillna(np.nan, inplace=True)
starter_df['Fully Vaccinated'].fillna(0, inplace=True)
starter_df['Daily Case Avg'].fillna(0, inplace=True)
#starter_df['Daily Case per 100k'].fillna(0, inplace=True)
```

```python
avg1 = starter_df['Total Deaths'].mean()
avg2 = starter_df['Total Deaths per 100k'].mean()

starter_df['Total Deaths'].fillna(avg1, inplace=True)
starter_df['Total Deaths per 100k'].fillna(avg2, inplace=True)
```

```python
starter_df['Daily Case Avg'] = starter_df['Daily Case Avg'].apply(lambda x: np.float64(str(x).replace(',','')))
#starter_df['Daily Case per 100k'].replace(to_replace = '<1', value = 0, inplace=True)
#starter_df['Daily Case per 100k'] = starter_df['Daily Case per 100k'].apply(lambda x: np.float64(str(x).replace(',','')))
```

```python
starter_df['Daily Case 14-Day change %'].fillna(np.nan, inplace=True)
starter_df['Daily Case 14-Day change %'].replace(to_replace = '<1', value = 0, inplace=True)
starter_df['Daily Case 14-Day change %'] = starter_df['Daily Case 14-Day change %'].apply(lambda x: np.float64(str(x).replace(',','')))
```

```python
starter_df['Deaths Daily Avg.'].fillna(np.nan, inplace=True)
starter_df['Deaths Daily Avg.'] = starter_df['Deaths Daily Avg.'].apply(lambda x: np.float64(str(x).replace(',','')))
```

```python
fv_mean = starter_df['Fully Vaccinated'].mean()
dca_mean = starter_df['Daily Case Avg'].mean()

starter_df['Fully Vaccinated'].fillna(fv_mean, inplace=True)
starter_df['Daily Case Avg'].fillna(dca_mean, inplace=True)
starter_df.head(50)
```
